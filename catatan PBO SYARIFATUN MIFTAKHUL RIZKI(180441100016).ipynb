# Inheritance
Inheritance merupakan konsep mewariskan attribute dan method yang dimiliki oleh sebuah class kepada class turunannya. 
Dengan konsep ini class yang dibuat cukup mendefinisikan attribute dan method yang spesifik didalamnya, sedangkan attribute dan method yang lebih umum akan didapatkan dari class yang menjadi induknya  
Subclass adalah class anak atau turunan dari superclass/parent class.








# Type of Inheritance
Pewarisan/Penurunan adalah konsep suatu pemrograman dimana sebuah class dapat menurunkan property dan method yang dimiliki kepada class lain.
Suatu kelas dapat mewarisi sebuah fungsi dari kelas lain.

1. DEFAULT

Default adalah tipe yang properti,properti dan method di dalam kelas induk sama persis dengan subclass menggunakan (pass)


2. OVERRIDING

Overriding adalah tipe inheritance yang kelas anak dapat memodifikasi salah satu method/properti dalam kelas induk tetapi kelas anak harus punya nama yang sama,tanda,dan parameter yang ada pada kelas induknya.


3. SUPER

Super adalah tipe inheritance mrngembalikan atribut dan method dari super objek (induk) suatu kelas (mengambil atribut dari induk)


4. MULTIPLE

Multiple adalah tipe inheritance memiliki dua atau lebih parent class, mewarisi semua properti



# Polymorphisme

1. Polymorphisme with a Function
Polymorphisme cara memanggilnya menggunakan fungsi




2. Polymorphisme with a Method
